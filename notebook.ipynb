```python
import numpy as np
np.random.seed(42)

def relu(z):
  return np.maximum(0,z)

def relu_derivatives(z):
  return (z>0).astype(float)

def sigmoid(z):
  return 1/(1+np.exp(-z))

w1= np.random.randn(2,4)*np.sqrt(2/2)
b1= np.zeros((1,4))*0.01

w2= np.random.randn(4,1)*np.sqrt(2/4)
b2= np.zeros((1,1))

def forward(X,w1,b1,w2,b2):
  z1=np.dot(X,w1)+b1

  a1=relu(z1)
  z2=np.dot(a1,w2)+b2

  a2=sigmoid(z2)
  return z1,a1,z2,a2

# now we will be computing losses
# once we compute the forward pass now its time for losses
def compute_losses(y_pred,y):
  epsilon = 1e-9
  y_pred= np.clip(y_pred,epsilon,1-epsilon)
  return -np.mean(
      y*np.log(y_pred)+
      (1-y)*np.log(1-y_pred)
  )

def backward(X, y, z1, a1, z2, a2, w2):
  m=X.shape[0]
  dz2=(a2-y)
  dw2=(1/m)*np.dot(a1.T,dz2)
  db2=(1/m)*np.sum(dz2,axis=0,keepdims=True)

  da1 = np.dot(dz2,w2.T)
  dz1 = da1*relu_derivatives(z1)
  dw1=(1/m)*np.dot(X.T,dz1)
  db1=(1/m)*np.sum(dz1,axis=0,keepdims=True)
  return dw1, db1,dw2,db2

def train(X,y,w1,b1,w2,b2, lr=0.5 ,epochs=5000):
  for i in range(epochs):
    z1, a1, z2, a2 = forward(X, w1, b1, w2, b2)

    loss = compute_losses(a2,y)
    dw1, db1, dw2, db2 = backward(X, y, z1, a1, z2, a2, w2)
    w1 -= lr * dw1
    b1 -= lr * db1
    w2 -= lr * dw2
    b2 -= lr * db2

    if i%100 ==0:
      print(f"epochs {i},loss:{loss}")
  return w1, b1, w2, b2

X=np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array([[0], [1], [1], [0]])
```

```python
w1,b1,w2,b2 = train(X,y,w1,b1,w2,b2)
```
